```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

# Structure factor

```python
s_k = pd.read_csv('Data/s_k.csv')
kpoints = s_k.k
s = s_k['s(k)']
```

```python
plt.plot(kpoints,s,c = 'red')
plt.xlabel('K')
plt.ylabel('S(k)')
plt.title('Structure Factor')

plt.savefig('Images/structurefactor.png',dpi = 300)
plt.show()
```

# Pair correlation

```python
g_r = pd.read_csv('Data/g_r.csv')
r = g_r.r
g = g_r['g(r)']
```

```python
plt.plot(r,g,c = 'red')
plt.xlabel(r'Distance ($\AA$)')
plt.ylabel('g(r)')
plt.title('Pair Correlation Function')

plt.savefig('Images/paircorr.png',dpi = 300)
plt.show()
```

# VV AC Time

```python
ac = pd.read_csv('Data/vcc_t.csv')
t = ac.t
#t = t[1:len(t)]
C = np.array(ac.C_vv)
C = C/2669
#C = C[1:len(C)]
```

```python
plt.plot(t,C,c = 'red')
plt.xlabel('Time (ps)')
plt.ylabel('Correlation')
plt.title('Velocity-Velocity Autocorrelation Time')
plt.xlim(0,0.02)

plt.savefig('Images/vv_actime.png',dpi = 300)
plt.show()
```

# Diffusion Coefficient

```python
meansq_dist = pd.read_csv('Data/r2_t.csv')
t = meansq_dist.t
r2 = meansq_dist['r2(t)']
```

```python
plt.plot(t,r2,c = 'red')
plt.xlabel('Time (ps)')
plt.ylabel('Mean Squared Displacement')
plt.title('Evaluation of Diffusion Coefficient')
plt.savefig('Images/diff.png',dpi = 300)
plt.show()
linregress(t, r2)
```
